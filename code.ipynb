```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle
import pandas as pd

training_data_path = 'Training.csv'
testing_data_path = 'Testing.csv'
training_df = pd.read_csv(training_data_path)
```

```python

X_train = training_df.drop('prognosis', axis=1)
y_train = training_df['prognosis']
```

```python
X_train
```

```python
testing_df = pd.read_csv(testing_data_path)
X_test = testing_df.drop('prognosis', axis=1)
y_test = testing_df['prognosis']

```

```python
# Check the column names and structure of both DataFrames
training_info = training_df.info()
testing_info = testing_df.info()

# Compare the columns in both DataFrames
training_columns = training_df.columns
testing_columns = testing_df.columns

training_columns, testing_columns
```

```python
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
X_train_split
```

```python
X_val_split
```

```python
y_train_split
```

```python
y_val_split
```

```python
X_train_split_df = pd.DataFrame(X_train_split, columns=X_train.columns)
X_val_split_df = pd.DataFrame(X_val_split, columns=X_train.columns)
y_train_split_df = pd.DataFrame(y_train_split, columns=['prognosis'])
y_val_split_df = pd.DataFrame(y_val_split, columns=['prognosis'])

# Save to CSV files
X_train_split_df.to_csv('X_train_split.csv', index=False)
X_val_split_df.to_csv('X_val_split.csv', index=False)
y_train_split_df.to_csv('y_train_split.csv', index=False)
y_val_split_df.to_csv('y_val_split.csv', index=False)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_split, y_train_split)
y_val_pred = rf_model.predict(X_val_split)
```

```python
rf_model
```

```python
y_val_pred
```

```python
validation_accuracy = accuracy_score(y_val_split, y_val_pred)
y_test_pred = rf_model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)
```

```python
test_accuracy
```

```python


# # Save the trained model using pickle
model_filename = 'healthcare_chatbot_model.pkl'
with open(model_filename, 'wb') as model_file:
    pickle.dump(rf_model, model_file)

# # Return validation accuracy, test accuracy, and model filename
```
